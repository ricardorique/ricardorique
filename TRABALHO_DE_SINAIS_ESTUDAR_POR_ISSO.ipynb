```python
#atividade
#CRIAR UM SINAL ARTIFICIAL
# PLOTAR O SINAL
#CALCULAR O FFT DO SINAL
#PLOTAR O FFT DO SINAL CORRETAMENTE
# selecionar celula sempre quando importar algo
import numpy as np #biblioteca que usei para criar o sinal artificial
import matplotlib.pyplot as plt #biblioteca que usei para plotar (criar o grafico)
import scipy# biblioteca que utilizei para colocar a transformada de fourier
```

Criando o Sinal Artificial


```python

freq = 10
tempo = np.linspace (0,1,1000)#a linha do tempo, o 0 é o inicio o 1 é o fim e o 1000 é a quantidade do t
fase = 20
amp = 25
```

```python
sinal = amp*np.sin(2*np.pi*freq*tempo + fase) #função seno amplitude x seno(2pi x frequencia x tempo +fase)
```

```python
sinal #printa os pontos do gráfico da função seno.
```

PLOTAR O GRAFICO

```python
plt.plot(sinal) #comando para criar o gráfico
```

CALCULO DE FFT (FOURIER) NO SINAL

```python
scipy.fft.fft(sinal)#coloca a transformada de fourier na função seno
```

```python
sinal_fft = scipy.fft.fft(sinal)# cria a variavel do sinal com fourier
```

PLOTAR SINAL COM FFT

```python
plt.plot(sinal_fft)#mostrar o desenho do gráfico
plt.show()#frescura pra deixar bonitinho.
```
